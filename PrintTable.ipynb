```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('output/db/db.sqlite')
conn.execute("ATTACH DATABASE 'data/db.sqlite' AS input_db")

df = pd.read_sql_query('''
    SELECT 
        r.repository,
        r.score_env, r.score_data, r.score_docs, 
        r.score_code, r.score_repro, r.score_total,
        'https://doi.org/' || a.doi AS paper_link
    FROM repositories r
    LEFT JOIN input_db.repositories ir ON r.repository = ir.repository
    LEFT JOIN input_db.article a ON ir.article_id = a.id
    ORDER BY r.score_total DESC
''', conn)

conn.close()
df
```
